```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
sns.set()
```

```python
elections = pd.read_csv("2016_US_County_Level_Presidential_Results.csv")
elections.head()
```

```python
_ = plt.hist(elections['per_dem'], bins=20)
_ = plt.xlabel('percent of vote for Clinton')
_ = plt.ylabel('number of counties')
plt.show()
```

```python
#Plotting all of the data with bee swarm plots

_ = sns.swarmplot(x='state_abbr', y='per_dem', data=elections)
_ = plt.xlabel('state')
_ = plt.ylabel('percent of vote for Clinton')
plt.margins(0.02) #Keeps data off plot edges
plt.show()
```

```python
x = np.sort(elections['per_dem'])
y = np.arange(1, len(x)+1 / len(x))
_ = plt.plot(x, y, marker='.', linestyle='none')
_ = plt.xlabel('percent of vote for Obama')
_ = plt.ylabel('ECDF')
plt.margins(0.02) #Keeps data off plot edges
plt.show()
```
